```python
tienda4['Fecha de Compra'] = pd.to_datetime(tienda4['Fecha de Compra'], dayfirst=True)

ingresos_por_fecha = (
    tienda4
    .groupby('Fecha de Compra')['Precio']
    .sum()
    .reset_index()
)
```

```python
plt.figure(figsize=(10,5))
plt.plot(
    ingresos_por_fecha['Fecha de Compra'],
    ingresos_por_fecha['Precio']
)
plt.title('Evolución de ingresos en el tiempo - Tienda 4')
plt.xlabel('Fecha de compra')
plt.ylabel('Ingresos')
plt.xticks(rotation=45)
plt.show()
```

# **Informe final - Tiendas Alura Store**

## **Introducción**

El objetivo de este análisis es identificar cuál de las cuatro tiendas de la cadena Alura Store presenta el menor nivel de eficiencia, con el fin de recomendar al Sr. Juan cuál debería vender para iniciar un nuevo emprendimiento.
Para ello, se analizaron distintos indicadores clave de desempeño, entre ellos: ingresos totales, categorías de productos más y menos vendidas, calificaciones promedio de los clientes, productos más y menos vendidos y el comportamiento de las ventas a lo largo del tiempo. Los resultados fueron respaldados mediante visualizaciones gráficas.


## **Análisis de resultados**

### **1. Ingresos totales por tiendas**

Los ingresos totales muestran diferencias claras entre las tiendas:

- Tienda 1: $1.150.880.400

- Tienda 2: $1.116.343.500

- Tienda 3: $1.098.019.600

- Tienda 4: $1.038.375.700

La Tienda 1 es la que mayor volumen de ingresos genera, mientras que la Tienda 4 presenta el ingreso total más bajo. Esta diferencia sugiere un menor desempeño financiero general de la Tienda 4 frente a las demás.


### **2. Categorías de productos más y menos vendidas**

En todas las tiendas, la categoría Muebles se posiciona como la más vendida, lo que indica una tendencia clara de preferencia por parte de los clientes.
El top 3 de categorías en todas las tiendas es:

- Muebles

- Electrónicos

- Juguetes

En contraste, las categorías menos vendidas varían:

Tienda 1 y 2: Artículos para el hogar

Tienda 3 y 4: Instrumentos musicales

Esto evidencia que la Tienda 4 comparte no solo la categoría menos vendida con la Tienda 3, sino también una menor diversificación en categorías de alto rendimiento.

### **3. Calificación promedio de los clientes**

Las calificaciones promedio son muy similares entre las tiendas:

- Tienda 1: 4.7

- Tienda 2: 4.7

- Tienda 3: 4.7

- Tienda 4: 4.7 (ligeramente inferior)

Aunque la diferencia es mínima, la Tienda 4 presenta la calificación promedio más baja, lo que puede indicar una percepción ligeramente menos favorable por parte de los clientes.

### **4. Productos más y menos vendidos**

Los productos más vendidos por tienda son:

- Tienda 1: Armario

- Tienda 2: Iniciando en programación

- Tienda 3: Kit de bancas

- Tienda 4: Cama box

Los productos menos vendidos son:

- Tienda 1: Auriculares con micrófono

- Tienda 2: Juego de mesa

- Tienda 3: Bloques de construcción

- Tienda 4: Guitarra eléctrica

En el caso de la Tienda 4, el producto menos vendido pertenece a la categoría de instrumentos musicales, que ya se había identificado como una de las categorías con peor desempeño, reforzando la idea de baja rotación en ciertos segmentos.

### **5. Comportamiento de ventas en el tiempo**

A partir del gráfico de líneas de ingresos a lo largo del tiempo, se observa que la Tienda 1 presenta mayores picos de ventas y un comportamiento más dinámico.
Por el contrario, la Tienda 4 muestra menos picos y un volumen de ventas más bajo de forma consistente, lo que indica una menor capacidad de generar ingresos en periodos clave.

### **6. Conclusión y recomendación final**

Con base en el análisis de todos los indicadores, se concluye que la Tienda 4 es la menos eficiente de la cadena Alura Store. Esta tienda presenta:

- El menor ingreso total entre las cuatro tiendas.

- Menor dinamismo en ventas a lo largo del tiempo.

- Categorías menos vendidas con bajo rendimiento (instrumentos musicales).

- Una calificación promedio ligeramente inferior a las demás.

Si bien mantiene buenas ventas en la categoría de muebles, su desempeño general es inferior en comparación con tiendas como la Tienda 1, que lidera en ingresos y presenta mayor estabilidad y picos de venta.

Se recomienda al Sr. Juan vender la Tienda 4, ya que es la que aporta menor valor relativo a la cadena y permitiría liberar recursos para invertir en un nuevo emprendimiento con mayor potencial de crecimiento.
