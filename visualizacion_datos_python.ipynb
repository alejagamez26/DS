# Visualización de datos con Pyhton: Matplotlib, Seaborn y Plotly


```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc["Colombia", anos]
colombia
```

```python
col_dict = {"Año":colombia.index.tolist(),
            "Inmigrantes":colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(["1980", "1985", "1990", "1995", "2000", "2005", "2010"])
plt.title("Inmigración de colombianos hacia Canadá")
plt.xlabel("Año")
plt.ylabel("Número de inmigrantes")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_xlabel("Año")
ax.set_ylabel("Número de Inmigrantes")
ax.set_title("Inmigración de colombianos hacia Canadá\n1980 - 2013")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,4))

ax[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
ax[0].set_title("Inmigración de Colombianos hacia Canadá\n1980 - 2013")
ax[0].set_xlabel("Año")
ax[0].set_ylabel("Número de Inmigrantes")
ax[0].xaxis.set_major_locator(plt.MultipleLocator(5))
ax[0].grid()

ax[1].boxplot(datos_col['Inmigrantes'])
ax[1].set_title("Distribución de la migración de colombianos hacia Canadá\n1980 - 2013")
ax[1].set_xlabel("Colombia")
ax[1].set_ylabel("Número de Inmigrantes")
ax[1].grid()

plt.show()
```

```python
datos_col.describe()
```

```python
fig, ax = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle("Inmigración de los 4 mayores países sudamericanos\n hacia Canadá 1980 - 2013")

ax[0,0].plot(df.loc["Colombia", anos])
ax[0,0].set_title("Colombia")

ax[0,1].plot(df.loc["Brasil", anos])
ax[0,1].set_title("Brasil")

ax[1,0].plot(df.loc["Argentina", anos])
ax[1,0].set_title("Argentina")

ax[1,1].plot(df.loc["Perú", anos])
ax[1,1].set_title("Perú")

ymin = 0
ymax = 7000

# Combine both loops into one and use a different loop variable to avoid overwriting 'ax'
for subplot_ax in ax.flat:
  subplot_ax.set_ylim(ymin, ymax)
  subplot_ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  subplot_ax.set_xlabel("Año")
  subplot_ax.set_ylabel("Número de Inmigrantes")
  subplot_ax.grid()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'], lw = 2, color = "g") #marker = "o"
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_xlabel("Año", fontsize = 14)
ax.set_ylabel("Número de Inmigrantes", fontsize = 14)
ax.set_title("Inmigración de colombianos hacia Canadá\n1980 - 2013", fontsize=18) #loc = "left"
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.grid(linestyle = "--")
plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values("Total", ascending = True)

colores = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'olive', 'cyan', 'magenta']

fig, ax = plt.subplots(figsize=(12,5))

ax.barh(sudamerica_sorted.index, sudamerica_sorted["Total"], color=colores)
ax.set_title("Inmigración sudamericanos a Canadá\n 1980 - 2013", fontsize = 18)
ax.set_ylabel("", fontsize = 14)
ax.set_xlabel("Número de Inmigrantes", fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == "Colombia":
    colores.append("forestgreen")
  else:
    colores.append("silver")

fig, ax = plt.subplots(figsize=(12,5))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis="y", left=False)

ax.barh(sudamerica_sorted.index, sudamerica_sorted["Total"], color=colores)
ax.set_title("Inmigración de colombianos hacia Canadá\n 1980 - 2013", fontsize = 18)
ax.set_ylabel("", fontsize = 14)
#ax.set_xlabel("Número de Inmigrantes", fontsize = 14)
#ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i,j in enumerate(sudamerica_sorted["Total"]):
  ax.text(j+20, i, str(j), color = "black", fontsize = 8.5, ha="left", va="center")

fig.savefig("/content/inmigrantes_sudamericanos_canada.png",
            transparent = False, dpi = 300, bbox_inches = "tight")
```

## Seaborn

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10 = df.sort_values("Total", ascending = False).head(10)
top_10
```

```python
ax = sns.barplot(data = top_10, x = "Total", y = top_10.index, orient = "h")
ax.set(title = "Top 10 países con mayor migración hacia Canadá\n1980 - 2013",
       xlabel = "Número de Inmigrantes",
       ylabel = "Año")
plt.show()
```

```python
fig.ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data = top_10, x = "Total", y = top_10.index, orient="h" )
ax.set_title("Top 10 países con mayor migración hacia Canadá\n1980 - 2013",
             loc="left", fontsize=16)
ax.set_xlabel("Número de Inmigrantes", fontsize = 14)
ax.set_ylabel("")
plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis="y", left=False)
  ax = sns.barplot(data = top_10, x = "Total", y=top_10.index, hue = top_10.index, orient="h",
                   palette=palette, legend = False)
  ax.set_title("Top 10 países con mayor migración hacia Canadá\n1980 - 2013",
              loc="left", fontsize=16)
  ax.set_xlabel("Número de Inmigrantes", fontsize = 14)
  ax.set_ylabel("")
  for i,j in enumerate(top_10["Total"]):
    ax.text(j+20, i, str(j), color = "black", fontsize = 8.5, ha="left", va="center")

  plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])
  ax = sns.barplot(data = top_10, x = "Total", y=top_10.index, hue = top_10.index, orient="h",
                   palette=palette, legend = False)
  ax.set_title("Top 10 países con mayor migración hacia Canadá\n1980 - 2013",
              loc="left", fontsize=16)
  ax.set_xlabel("Número de Inmigrantes", fontsize = 14)
  ax.set_ylabel("")
  sns.despine()
  for i,j in enumerate(top_10["Total"]):
    ax.text(j+20, i, str(j), color = "black", fontsize = 8.5, ha="left", va="center")

  plt.show()
```

```python
generar_grafico("tab10")
```

## Plotly

```python
import plotly.express as px
```

```python
fig = px.line(datos_col, x = "Año", y = "Inmigrantes",
              title="Inmigración de colombianos hacia Canadá\n1980 - 2013")
fig.update_traces(line_color="green", line_width=3)
fig.update_layout(width=800, height=400, xaxis = {"tickangle":-45},
                  xaxis_title="Año", yaxis_title="Número de Inmigrantes",
                  font_family="Arial", font_size=14, font_color="grey")
fig.show()
```
